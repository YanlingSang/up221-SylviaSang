## Explore The Relationship Between Vehicle Ownership & Bicycle Crash Rate  
I processed and cleaned Census Tract, vehicle ownership, and bicycle crash datasets, ensuring standardized column names and coordinate systems. I converted crash data into a geospatial format and assigned them to Census Tracts. I analyzed the data by calculating total crashes, crash rates per 1,000 people and vehicles, and performed correlation and regression analysis to explore the relationship between vehicle ownership and bicycle crash rates. Finally, I created both a static map (Matplotlib) and an interactive map (Folium) to visualize vehicle ownership distribution and crash locations.

# Part 1: Data Reading & Processing

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import numpy as np
import folium
import mapclassify
import seaborn as sns

# Load Census Tracts geographic data
census_gdf = gpd.read_file("2020_Census_Tracts.shp")

# Load vehicle ownership data
vehicles_df = pd.read_csv("new_vehicles_dataset.csv")

# Load bicycle crash data
crashes_df = pd.read_csv("23_Bicycle_Crashes.csv")

# View data structure
print("Census Tracts Data:", census_gdf.head())
print("Vehicle Data:", vehicles_df.head())
print("Crash Data:", crashes_df.head())
```

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Point  # ✅ Import Point to fix the error

# Ensure Census Tract ID is consistent
print("Census data column names:", census_gdf.columns)
print("Crash data column names:", crashes_df.columns)

# The Census Tract ID column may not be named 'Census_Tract', so we need to check
if 'LABEL' in census_gdf.columns:
    census_gdf = census_gdf.rename(columns={'LABEL': 'Census_Tract'})

# Standardize coordinate reference system (CRS)
print("Original Census Tracts CRS:", census_gdf.crs)
if census_gdf.crs != "EPSG:4326":
    census_gdf = census_gdf.to_crs(epsg=4326)
print("Transformed Census Tracts CRS:", census_gdf.crs)

# Convert crash data to geographic format
crashes_df = crashes_df.dropna(subset=["POINT_X", "POINT_Y"])  # Remove rows without coordinates
geometry = [Point(xy) for xy in zip(crashes_df["POINT_X"], crashes_df["POINT_Y"])]  # ✅ FIXED: Now Point is recognized
crashes_gdf = gpd.GeoDataFrame(crashes_df, geometry=geometry, crs="EPSG:4326")

# Perform spatial join
crashes_gdf = crashes_gdf.sjoin(census_gdf, how="left", predicate="within")

# Verify join results
print("Sample of joined crash data:", crashes_gdf.head())

# Calculate the number of crashes per Census Tract
if 'Census_Tract' in crashes_gdf.columns:
    crash_counts = crashes_gdf.groupby("Census_Tract").size().reset_index(name="Total_Crashes")
else:
    print("Error: Spatial join did not correctly assign Census Tract. Please check the data!")

# Merge crash statistics back into Census data
census_gdf = census_gdf.merge(crash_counts, on="Census_Tract", how="left")

# Fill in areas with no crashes
census_gdf["Total_Crashes"] = census_gdf["Total_Crashes"].fillna(0)

# Final verification
print("Final merged data:", census_gdf.head())
```

```python
# Remove duplicate Total_Crashes columns
census_gdf = census_gdf.drop(columns=["Total_Crashes_x", "Total_Crashes_y"], errors='ignore')

# Ensure that empty Total_Crashes values are filled with 0
census_gdf["Total_Crashes"] = census_gdf["Total_Crashes"].fillna(0)

# View the cleaned data
print(census_gdf.head())
```

```python
# Ensure all data is numeric
census_gdf["Total_Crashes"] = pd.to_numeric(census_gdf["Total_Crashes"], errors="coerce")
census_gdf["ShapeSTAre"] = pd.to_numeric(census_gdf["ShapeSTAre"], errors="coerce")
census_gdf["CT20"] = pd.to_numeric(census_gdf["CT20"], errors="coerce")

# Fill NaN values to avoid calculation errors
census_gdf[["Total_Crashes", "ShapeSTAre", "CT20"]] = census_gdf[
    ["Total_Crashes", "ShapeSTAre", "CT20"]
].fillna(0)

# Calculate crash rate per 1,000 people
census_gdf["crash_rate_per_1000"] = (census_gdf["Total_Crashes"] / census_gdf["ShapeSTAre"]) * 1000

# Calculate crash rate per 1,000 vehicles
census_gdf["crash_rate_per_1000_vehicles"] = (census_gdf["Total_Crashes"] / census_gdf["CT20"]) * 1000

# Ensure all NaN values are replaced with 0
census_gdf[["crash_rate_per_1000", "crash_rate_per_1000_vehicles"]] = census_gdf[
    ["crash_rate_per_1000", "crash_rate_per_1000_vehicles"]
].fillna(0)

# Print the first few rows for verification
print(census_gdf[["Census_Tract", "Total_Crashes", "crash_rate_per_1000", "crash_rate_per_1000_vehicles"]].head())
```

# part2 Correlation Analysis

```python
print("Correlation Analysis:")
print(census_gdf[["CT20", "crash_rate_per_1000", "crash_rate_per_1000_vehicles"]].corr())
```

```python
# Define X and Y
X = census_gdf[['CT20']]  # Vehicle ownership
y = census_gdf['crash_rate_per_1000']  # Crash rate per 1,000 people
X = sm.add_constant(X)  # Add intercept term

# Linear regression model
model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
# Apply logarithm to prevent data skewness
census_gdf["log_CT20"] = np.log(census_gdf["CT20"] + 1)

X_log = sm.add_constant(census_gdf[['log_CT20']])
log_model = sm.OLS(y, X_log).fit()
print(log_model.summary())
```

```python
# Select top 10 Census Tracts with the highest crash rate per 1,000 people
top_crash_tracts = census_gdf.nlargest(10, "crash_rate_per_1000")

# Set figure size
plt.figure(figsize=(12, 6))

# Create bar chart (explicitly setting hue to avoid warning)
sns.barplot(
    data=top_crash_tracts,
    x="Census_Tract",
    y="crash_rate_per_1000",
    hue="Census_Tract",  # Assign hue to x variable
    palette="Reds",
    legend=False  # Remove redundant legend
)

# Labels and title
plt.xlabel("Census Tract")
plt.ylabel("Crash Rate per 1,000 People")
plt.title("Top 10 Census Tracts with Highest Bicycle Crash Rates")

# Rotate x-axis labels for readability
plt.xticks(rotation=45)

# Show plot
plt.show()
```

```python
# Set figure size
plt.figure(figsize=(10, 6))

# Scatter plot with regression line
sns.regplot(
    data=census_gdf,
    x="CT20",
    y="crash_rate_per_1000",
    scatter_kws={'alpha':0.5},  # Set transparency for points
    line_kws={'color':'red'},   # Regression line color
)

# Labels and title
plt.xlabel("Vehicle Ownership per 1,000 People")
plt.ylabel("Bicycle Crash Rate per 1,000 People")
plt.title("Relationship Between Vehicle Ownership and Bicycle Crash Rate")

# Show plot
plt.show()
```

```python
# Compute correlation matrix
corr_matrix = census_gdf[["CT20", "crash_rate_per_1000", "crash_rate_per_1000_vehicles"]].corr()

# Set figure size
plt.figure(figsize=(8, 6))

# Create heatmap
sns.heatmap(
    corr_matrix, 
    annot=True,  # Display correlation values
    cmap="coolwarm",  # Color scheme
    linewidths=0.5,  # Add lines between cells
    fmt=".2f"  # Format decimal places
)

# Title
plt.title("Correlation Heatmap: Vehicle Ownership & Crash Rate")

# Show plot
plt.show()
```

# part 3 Map Visualization

```python
print("Census Tract column names:", census_gdf.columns)
if 'LABEL' in census_gdf.columns:
    census_gdf.rename(columns={'LABEL': 'Census_Tract'}, inplace=True)
census_gdf = census_gdf[~census_gdf["geometry"].isna()]
```

```python
# Create map
m = folium.Map(location=[34.05, -118.25], zoom_start=10)

# Add base layer
folium.Choropleth(
    geo_data=census_gdf,
    data=census_gdf,
    columns=["Census_Tract", "CT20"],
    key_on="feature.properties.Census_Tract",
    fill_color="YlGnBu",
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name="Vehicle Ownership per 1,000 People"
).add_to(m)

# Add crash rate markers
for _, row in census_gdf.iterrows():
    if row["crash_rate_per_1000"] > 0 and row["geometry"].centroid.is_valid:
        folium.CircleMarker(
            location=[row["geometry"].centroid.y, row["geometry"].centroid.x], 
            radius=row["crash_rate_per_1000"] * 5,
            color="red",
            fill=True,
            fill_color="red",
            fill_opacity=0.6,
            popup=f"Census Tract: {row['Census_Tract']}<br>Crash Rate: {row['crash_rate_per_1000']:.2f}"
        ).add_to(m)

# Save the map
m.save("Ownership_vs_Crash_Map.html")
print("Map successfully generated!")
```

```python
# Remove `index_right` to avoid conflicts
for df in [crashes_gdf, census_gdf]:
    if "index_right" in df.columns:
        df.drop(columns=["index_right"], inplace=True)

print("`index_right` column removed to prevent sjoin() conflicts.")
print("Census_gdf geometry type:", census_gdf.geom_type.unique())
print("Crashes_gdf geometry type:", crashes_gdf.geom_type.unique())

# Remove NaN geometry data
census_gdf = census_gdf[~census_gdf["geometry"].isna()]
crashes_gdf = crashes_gdf[~crashes_gdf["geometry"].isna()]

print(f"After filtering, census_gdf contains {len(census_gdf)} records, crashes_gdf contains {len(crashes_gdf)} records.")

import geopandas as gpd
from shapely.geometry import Point

# Recreate geometry column for crash data
crashes_gdf["geometry"] = [Point(xy) for xy in zip(crashes_gdf["POINT_X"], crashes_gdf["POINT_Y"])]
crashes_gdf = gpd.GeoDataFrame(crashes_gdf, geometry="geometry", crs="EPSG:4326")  # Ensure EPSG:4326

# Standardize Census Tracts coordinate system
if census_gdf.crs != "EPSG:4326":
    census_gdf = census_gdf.to_crs(epsg=4326)

# Ensure `index_right` is removed
for df in [crashes_gdf, census_gdf]:
    if "index_right" in df.columns:
        df.drop(columns=["index_right"], inplace=True)

# Perform spatial join
crashes_gdf = crashes_gdf.sjoin(census_gdf, how="left", predicate="within")

# Check if `Census_Tract` was successfully matched
if "Census_Tract" in crashes_gdf.columns:
    print("`Census_Tract` successfully matched.")
else:
    print("`Census_Tract` is still missing. Please check if `census_gdf` contains the correct `geometry` data.")
```

```python
# Create Matplotlib figure
fig, ax = plt.subplots(figsize=(12, 12))  # Adjust size to prevent the map from being too small

# Ensure data is numeric
census_gdf["CT20"] = pd.to_numeric(census_gdf["CT20"], errors="coerce")

# Select color scheme
color_scheme = "YlGnBu"  # Color mapping, options include 'coolwarm', 'viridis', 'Blues'

# Plot Census Tracts with color representing vehicle ownership rate
census_gdf.plot(
    column="CT20",  
    cmap=color_scheme,  
    linewidth=0.5,
    edgecolor="black",
    legend=True,
    legend_kwds={"title": "Vehicle Ownership Rate per 1,000 People"},  # Correct legend parameter
    scheme="quantiles",  # Ensure even color distribution
    ax=ax
)

# Plot crash points (red)
crashes_gdf.plot(
    ax=ax,
    markersize=5,
    color="red",
    alpha=0.6,
    label="Bicycle Crashes"
)

# Ensure proper map size
ax.set_xlim(census_gdf.total_bounds[0], census_gdf.total_bounds[2])
ax.set_ylim(census_gdf.total_bounds[1], census_gdf.total_bounds[3])

# Prevent map distortion
ax.set_aspect(1.5)  # Maintain correct aspect ratio

# Add title
plt.title("Vehicle Ownership & Bicycle Crash Rate by Census Tract", fontsize=16)

# Adjust legend
plt.legend(["Bicycle Crashes"], loc="upper right")  # Add legend only for crash points

# Remove axis
plt.axis("off")

# Save high-quality image
plt.savefig("Vehicle_Ownership_Crash_Map.png", dpi=300, bbox_inches="tight")

# Show map
plt.show()
```

### Key Findings:
1. Bicycle crashes are concentrated in urban areas, particularly where vehicle ownership is lower.

2. Higher vehicle ownership correlates with lower crash rates per capita, but crash counts remain high in dense areas.

3. Policy Recommendation: Improve bike infrastructure and road safety in high-risk, high-crash areas.
